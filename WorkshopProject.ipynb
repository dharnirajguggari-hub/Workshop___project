```python
#How has startup funding trend changed over the years?

import matplotlib.pyplot as plt
demo = data.groupby("year")["amount"].sum()
plt.figure(figsize=(10,5))
plt.plot(demo.index, demo.values,marker="o",markerfacecolor="red",color="black",markeredgecolor="green",linestyle="--",linewidth=1.5,label="Total Amount")
plt.title("Startup Funding Trend Over  Years",color="blue")
plt.xlabel("Year",color="blue")
plt.ylabel("Total Amount")
plt.grid()
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

Demo = data["city"].value_counts().head(10)

color = ["red","blue","purple","green","black","pink","orange","skyblue","magenta","yellow"]

plt.figure(figsize=(10,5))
plt.bar(Demo.index, Demo.values, color=color,alpha=0.4,edgecolor="red")

plt.xlabel("City", color="green")
plt.ylabel("Number of Startups", color="blue")
plt.title("Top 10 Startup Cities in India", color="red")

plt.xticks(rotation=90)
plt.grid(axis="y")
plt.show()
```

```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt

Demo = data["vertical"].value_counts().head(10)

color = ["red","black","pink","orange","green",
         "brown","yellow","purple","skyblue","blue"]

plt.figure(figsize=(10,5))
plt.barh(Demo.index[::-1], Demo.values[::-1], color=color,alpha=0.6,edgecolor="black")

plt.title("Top 10 Startup Sectors",color="red")
plt.xlabel("Number of Startups",color="green")
plt.ylabel("Sector",color="blue")
plt.legend()
plt.grid(axis="x",linestyle="--",color="grey")
plt.show()
```

```python
#What is the distributions of different funding rounds?
import matplotlib.pyplot as plt

Demo = data["round"].value_counts().head(7)

plt.figure(figsize=(6,8))
plt.pie(Demo.values,labels=Demo.index,autopct="%1.1f%%")

plt.title("Rounds distribution in percentage wise")
plt.show()
```

```python
#How are startup funding amount distributed?
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.hist(data["amount"], bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution of startup funding",color="red")


plt.show()
```

```python
#The relationship between funding year and funding amount?
import matplotlib.pyplot as plt

plt.figure(figsize=(7,6))
plt.scatter(data["year"], data["amount"], color="blue")

plt.title("Funding Amount by year",color="blue")
plt.xlabel(" Year",color="green")
plt.ylabel(" Amount",color="red")
plt.grid(linestyle="--")

plt.show()
```

```python
#Which funding round occurs most frequently?usng count plot using round column
import seaborn as sns
Demo = data["round"].value_counts().head(10).index
plt.figure(figsize=(9,5))
sns.countplot(data, y="round",order=Demo, palette="viridis")
plt.title("Funding Rounds Frequency")
plt.xlabel("Funding Round")
plt.ylabel("Count")

plt.show()
```
